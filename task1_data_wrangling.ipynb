```python
df = df.dropna(subset=["Satisfaction Level"])
```

```python
df['Age'].describe()
```

# Feature Engineering

```python
df['Spend Category'] = pd.qcut(df['Total Spend'], 3, labels=["Low", "Medium", "High"])

df['Recency Category'] = pd.cut(
    df['Days Since Last Purchase'],
    bins=[0, 30, 90, 365],
    labels=["Recent", "Moderate", "Inactive"]
)

df['Revenue per Item'] = df['Total Spend'] / df['Items Purchased']
```
